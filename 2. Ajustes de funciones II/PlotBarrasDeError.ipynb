# Grafico de datos con Barras de Error

Gr√°ficos mostrando uso de diferentes formas de Barras de Error  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# example variable error bar values
y_err = 0.1 + 0.2*np.sqrt(y)
x_err = 0.1 + 0.2*np.sqrt(x)
```

```python
np.set_printoptions(precision=4, suppress=False)
# If suppress is False, then scientific notation is used when absolute value of the smallest 
# number is < 1e-4 or the ratio of the maximum absolute value to the minimum is > 1e3.
print('x: ', x)
print('y: ', y)
print('..................................................................')
print('xerr: ', x_err)
print('yerr: ', y_err)
np.set_printoptions()  # formatter gets reset
```

```python
# First illustrate basic pyplot interface, using defaults where possible.
plt.errorbar(x, y, xerr=0.2, yerr=0.3, label='$y = exp(-x)$')
plt.title("Simplest errorbars, 0.2 in x, 0.4 in y")
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.grid()
```

```python
# plot with symmetric vertical error bar, symmetric -> side by side
plt.errorbar(x, y, yerr=y_err, fmt='o', label='$y = exp(-x)$')
plt.title('Vert. symmetric')
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.grid()
```

```python
# plot with symmetric horizontal error bar, symmetric -> side by side
plt.errorbar(x, y, xerr=x_err, fmt='o', label='$y = exp(-x)$')
plt.title('Hor. symmetric')
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.grid()
```

```python
# plot with asymmetric horizontal and vertical error bar, asymmetric -> side by side
plt.errorbar(x, y, xerr=[x_err, 2*x_err], yerr=[y_err, 2*y_err], fmt='o', label='$y = exp(-x)$')
plt.title('H & V asymmetric')
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.grid()
```

```python
# plot with symmetric horizontal and vertical error bar and logaritimic scale
plt.errorbar(x, y, xerr=[x_err, 2*x_err], yerr=[y_err, 2*y_err], fmt='o', label='$y = exp(-x)$')
plt.title('Mixed sym., log y')
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.yscale('log')
plt.grid()
```

```python
# First illustrate pyplot interface, using capsize.
# plt.figure()
plt.errorbar(x, y, fmt='o', xerr=0.2, yerr=0.4, capsize = 3, label='$y = exp(-x)$')
plt.title("Simplest errorbars, 0.2 in x, 0.4 in y")
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.grid()
```

```python
# Illustrate pyplot interface, using color errorbars.
plt.errorbar(x, y, y_err, marker='s', mfc='red', mec='green', ms=10, mew=2, label='$y = exp(-x)$')
# markerfacecolor (mfc), markeredgecolor (mec), markersize (ms) and markeredgewidth (mew).

plt.title("Color errorbars")
plt.xlabel('x [a.u.]')
plt.ylabel('y(x) [a.u.]')
plt.legend()
plt.grid()
```
