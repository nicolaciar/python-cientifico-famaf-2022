# Polinomios - Hermite Series, “Physicists” 
### (numpy.polynomial.hermite)

Este módulo proporciona una serie de objetos (principalmente funciones) útiles para trata de la serie Hermite, que incluye una Hermiteclase que encapsula las operaciones aritméticas usuales.  
Información general sobre cómo este módulo representa y trabaja con tales polinomios está en el docstring para su subpaquete "principal", numpy.polynomial. 

```python
import numpy as np
from numpy.polynomial import hermite as H  # importamos la librería y la llamamos H
```

```python
import matplotlib.pyplot as plt             # Importamos matplotlib.pyplot como el alias plt
```

```python
# Generamos las coordenadas en x
x = np.linspace(-1,1,51)

# Display the x-coordinate
print("X Co-ordinate...\n",x)
```

```python
# Creamos la variable y, en este caso proponemos una función cúbica con ruido
y = x**3 - x + 0.1* np.random.randn(len(x))  # el random agrega ruido
print("\nY Co-ordinate...\n",y)
```

```python
# Creamos la figura
plt.figure()

# Dibujamos los datos experimentales.
plt.plot(x,y,'ro', label = 'Experimental')

# Añadimos la leyenda
plt.legend()

plt.xlabel('Variable Indep (u.a.)')
plt.ylabel('Experimento (u.a.)')
plt.grid()
#plt.grid(b=True, color='g')
# Mostramos la gráfica
plt.show()
```

```python
c1 = H.hermfit(x, y, 1)
c2 = H.hermfit(x, y, 2)
c = H.hermfit(x, y, 3)  # Ajustamos los datos con un polinomio 3
#  Podemos probar con distintos grado, 1, 2, 3
```

```python
print(c1)
print(c2)
print(c)
```

```python
# Creamos la figura
plt.figure()

# Dibujamos los datos experimentales.
plt.plot(x,y,'ro', label = 'Experimental')

# Dibujamos
plt.plot(x,H.hermval(x, c1), 'b-', label= 'Fit 1')
plt.plot(x,H.hermval(x, c2), 'c-', label= 'Fit 2')
plt.plot(x,H.hermval(x, c), 'g-', label= 'Fit 3')
# Añadimos la leyenda
plt.legend()

plt.xlabel('Variable Indep (u.a.)')
plt.ylabel('Experimento (u.a.)')
plt.grid()
#plt.grid(b=True, color='g')
# Mostramos la gráfica
plt.show()
```

## Algebra con polinomios

```python
x = np.linspace(-4,4,21)
poliH_1 = H.Hermite.basis(3)(x)
poliH_2 = H.Hermite.basis(4)(x)
poliSumH = H.hermadd(poliH_1,poliH_2)
poliSubH = H.hermsub(poliH_1,poliH_2)

plt.plot(x, poliH_1, label='H 3')
plt.plot(x, poliH_2, label='H 4')
plt.plot(x, poliSumH, label ='Suma')
plt.plot(x, poliSubH, label ='Resta')
plt.legend()
plt.xlabel('x')
plt.ylabel('H(x)')
plt.grid()
plt.show()
```

```python
poliMulH = H.hermmul(poliH_1,poliH_2)
poliDivH = H.hermdiv(poliH_1,poliH_2)
```

```python
print(poliMulH)
```

```python
print(poliDivH)
```

```python
x = np.linspace(-1, 6)
CrootH = H.hermfromroots((1.0, 3.0 , 4.5))
qRootsH = H.hermval(x, CrootH )
#print(qRootsH)
#plt.plot(x, qRootsH)
plt.plot(x,  H.hermval(x, CrootH ))
plt.plot([-4, 7],[0, 0],'-k')
plt.plot([1, 1],[-10, 10],'-k')
plt.plot([3, 3],[-10, 10],'-k')
plt.plot([4.5, 4.5],[-10, 10],'-k')
plt.xlim(0, 5)
plt.ylim(-10,10)
plt.grid()
```
