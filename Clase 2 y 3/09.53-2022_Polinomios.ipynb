# Polinomios

Polynomials
Polynomials in NumPy can be created, manipulated, and even fitted using the convenience classes of the numpy.polynomial package, introduced in NumPy 1.4.

Documentation pertaining to specific functions defined for each kind of polynomial individually can be found in the corresponding module documentation:

    Power Series (numpy.polynomial.polynomial)
    Chebyshev Series (numpy.polynomial.chebyshev)
    Hermite Series, “Physicists” (numpy.polynomial.hermite)
    HermiteE Series, “Probabilists” (numpy.polynomial.hermite_e)
    Laguerre Series (numpy.polynomial.laguerre)
    Legendre Series (numpy.polynomial.legendre)
    Polyutils



```python
import numpy as np
from numpy.polynomial import polynomial as Poly
```

```python
import matplotlib.pyplot as plt             # Importamos matplotlib.pyplot como el alias plt
```

```python
np.set_printoptions(precision=4, suppress=False)
```

```python
# The x-coordinate
x = np.linspace(-1,1,51)

# Display the x-coordinate
print("X Co-ordinate...\n",x)
```

```python
# The y-coordinate
y = x**3 - x + 0.05* np.random.randn(len(x))
print("\nY Co-ordinate...\n",y)
```

```python
plt.plot(x, y, 'o', label='data')
#plt.plot(x, x**3 - x, label='true')
plt.grid()
plt.show()
```

```python
# To get the Least squares fit of Hermite series to data, use the hermite.hermfit() method in Python numpy

c, info = Poly.polyfit(x,y, 3, full=True)

print("\nResult...\n",c)

print("\nResult...\n",info)
# Display the array
print("Our Array...\n",c)

# Check the Dimensions
print("\nDimensions of our Array...\n",c.ndim)

# Get the Datatype
print("\nDatatype of our Array object...\n",c.dtype)

# Get the Shape
print("\nShape of our Array object...\n",c.shape)
```

```python
print(Poly.polyval(2.3, c))
```

```python
# Creamos la figura
plt.figure()

# Dibujamos los datos experimentales.
plt.plot(x, y,'ro', label = 'Experimental')
# Dibujamos
plt.plot(x, x**3 - x, label='true')
plt.plot(x, Poly.polyval(x, c), 'g-', label= 'Fit')
# Añadimos la leyenda
plt.legend()

plt.xlabel('Variable Indep (u.a.)')
plt.ylabel('Experimento (u.a.)')
plt.grid()
#plt.grid(b=True, color='g')
# Mostramos la gráfica
plt.show()
```

```python
q_Poly = Poly.Polynomial(c)
print(q_Poly)
```

```python
q6_poly = Poly.Polynomial([0, 0, 0, 1, 0, 0, 1])
print(q6_poly)
```

```python
q6_int = Poly.polyint(q6_poly)
print(q6_int)
print(Poly.Polynomial(q6_int))
```

```python
q6_int_der = Poly.polyder(q6_int)
print(Poly.Polynomial(q6_int_der))
```

```python
# Create an array of coefficients
k = np.array([1,2,3,4])

# Display the array
print("Our Array...\n",k)

# Check the Dimensions
print("\nDimensions of our Array...\n",k.ndim)

# Get the Datatype
print("\nDatatype of our Array object...\n",k.dtype)

# Get the Shape
print("\nShape of our Array object...\n",k.shape)

# To differentiate a Hermite series, use the hermite.hermder() method in Python
print("\nResult...\n",Poly.polyder(k))
```
