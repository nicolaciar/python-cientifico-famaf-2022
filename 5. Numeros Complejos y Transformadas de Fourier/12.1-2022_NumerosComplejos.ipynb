# Números complejos

Python por defecto interpreta los números complejos cuando usamos la letra j para denotar la parte imaginaria. Podemos entonces crear números complejos escribiendo el número directamente:

```python
z1 = 4 + 1j
z2 = 2-2j
z3 = -3j
```

```python
print("Número complejo z1: ",z1)
print("Número complejo z1: ",z2)
print("Número complejo z1: ",z3)
```

```python
print("Parte imaginaria", z1.imag)
print("Parte real", z2.real)
```

## Veamos las operaciones básicas con números complejos

### suma

```python
z1 + z2
```

### resta

```python
z2 - z1
```

### conjugado

```python
z1 = -1 + 3j
z1c = z1.conjugate()
print("Mi número complejo:",z1)
print("Mi número conjugado:",z1c)
```

### multiplicación

```python
z1*z2
```

### división

```python
z1/z2
```

### potencia

```python
z3**3
```

### raíz

```python
z3**(1/2)
```

## Libreria `cmath`
### Funciones matemáticas para números complejos

Este módulo proporciona acceso a funciones matemáticas para números complejos. Las funciones de este módulo aceptan enteros, números de coma flotante o números complejos como argumentos.  
También aceptarán cualquier objeto de Python que tenga un método `__complex __()` o `__float __()`: estos métodos se utilizan para convertir el objeto en un número complejo o de punto flotante, respectivamente, y luego la función se aplica al resultado de la conversión.
https://docs.python.org/3/library/cmath.html


```python
import cmath
```

### Conversión binómica/polar

```python
z1 = -1 + 3j
z1c = z1.conjugate()
print(z1, z1c)
modulo = abs(cmath.sqrt(z1*z1c))
print('modulo :', modulo)
print('parte real e imaginaria :', z1.real, z1.imag)
theta = cmath.atan(z1.imag/z1.real)+cmath.pi
print('theta :', abs(theta))

magnitud, angulo = cmath.polar(z1)
print('magnitud :', magnitud, 'ángulo :', angulo)
```

```python
z4 = complex(-1.0, 0.0)
z5 = -1+0j
z6 =  1+1j
z4ph = cmath.phase(z4)
z5ph = cmath.phase(z5)
z6ph = cmath.phase(z6)
print(abs(z4), z4ph)
print(cmath.polar(z4))
print(z5ph)
print(z6ph*180/cmath.pi)
```

```python
z6ph = cmath.phase(z6)
print(abs(z6), z6ph)
print(cmath.polar(z6))
```

### Polar a binómica

```python
magnitud, angulo = cmath.polar(z6)
print("z6:", z6)
print("Magnitud:", magnitud)
print("Angulo:", angulo*180/cmath.pi)
```

```python
z = cmath.rect(magnitud, angulo)
print("z=",z)
```

### Gráfica polar de números complejos

```python
import matplotlib.pyplot as plt
```

```python
z1 = 1 + 1j
print('z1 : ', z1)

magnitud, angulo = cmath.polar(z1)
print("Magnitud:", magnitud)
print("Angulo:", angulo*180/cmath.pi)
plt.polar([0,angulo],[0,magnitud],color='b',marker='o')
```
